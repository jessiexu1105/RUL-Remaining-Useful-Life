```python
def numPrimeArrangements( n: int) -> int:
        ctr = 0
    
        for num in range(n):
            if num <= 1:
                continue
            for i in range(2, num):
                if (num % i) == 0:
                    break
            else:
                ctr += 1

        return ctr
```

### Feature Engineering Steps
1. piece-wise RUL   
more likely to prevent the algorithm from overestimating the RUL; more logical model as the degradation of the system typically only starts after a certain degree of usage  

```python
def Piecewise_RUL(df, percentage):
    
    piecewise = []
    for id in df['machineID'].unique():
        max_rul=round(df.loc[(df['machineID'] == id) & (df['failed'] == 1)][['RUL']].mean()[0]*percentage)
        for i in df.loc[(df['machineID'] == id) & (df['failed'] == 1)].index:  
            stop = int(df[df.index == i]['RUL'])
            knee_point=stop-max_rul
            i_ini = i - stop + 1

            cycle_list = df.loc[(df.index >= i_ini) & (df.index <= i)]['RUL'].values
            
            piecewise_rul = []
            for j in range(len(cycle_list)):
                if knee_point<=0:
                    piecewise_rul.append(stop-cycle_list[j])
                elif j < knee_point:
                    piecewise_rul.append(max_rul)
                else:
                    rest = piecewise_rul[-1]-1
                    piecewise_rul.append(rest)

            piecewise.extend(piecewise_rul)

    return piecewise
```

```python
kaggle_train['RUL_piecewise']=Piecewise_RUL(kaggle_train, 0.75)
kaggle_test['RUL_piecewise']=Piecewise_RUL(kaggle_test, 0.75)
```

### 1. Piecewise vs. linear RUL

```python
df_piecewise_columns=['voltmean_24h',
       'rotatemean_24h', 'pressuremean_24h', 'vibrationmean_24h', 'voltsd_24h',
       'rotatesd_24h', 'pressuresd_24h', 'vibrationsd_24h','model', 'age', 'RUL_piecewise']
df_linear_columns=['voltmean_24h',
       'rotatemean_24h', 'pressuremean_24h', 'vibrationmean_24h', 'voltsd_24h',
       'rotatesd_24h', 'pressuresd_24h', 'vibrationsd_24h','model', 'age', 'RUL_I']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics 
from sklearn import preprocessing

# One Hot Encode
df_piecewise=pd.get_dummies(data=kaggle_train[df_piecewise_columns], columns=['model'])
df_linear=pd.get_dummies(data=kaggle_train[df_linear_columns], columns=['model'])

df_piecewise_test=pd.get_dummies(data=kaggle_test[df_piecewise_columns], columns=['model'])
df_linear_test=pd.get_dummies(data=kaggle_test[df_linear_columns], columns=['model'])

# X, y
X_train_piecewise = df_piecewise.drop(['RUL_piecewise'], axis=1)
y_train_piecewise = df_piecewise.RUL_piecewise
X_test_piecewise = df_piecewise_test.drop(['RUL_piecewise'], axis=1)
y_test_piecewise = df_piecewise_test.RUL_piecewise

X_train_linear=df_linear.drop(['RUL_I'], axis=1)
y_train_linear = df_linear.RUL_I
X_test_linear=df_linear_test.drop(['RUL_I'], axis=1)
y_test_linear = df_linear_test.RUL_I
```

```python
from sklearn.preprocessing import StandardScaler 
scaler_piecewise = StandardScaler()  
X_train_piecewise= scaler_piecewise.fit_transform(X_train_piecewise)  
X_test_piecewise = scaler_piecewise.transform(X_test_piecewise)

scaler_linear = StandardScaler()  
X_train_linear = scaler_linear.fit_transform(X_train_linear)  
X_test_linear = scaler_linear.transform(X_test_linear)
```

```python
from sklearn.metrics import mean_squared_error as mse
rf_model=RandomForestRegressor(n_estimators=500, min_samples_split=10, min_samples_leaf=4, max_depth=80, random_state=1)

rf_model.fit(X_train_piecewise, y_train_piecewise)
prediction=rf_model.predict(X_test_piecewise)
print('The R square of the Random Forests on train set is', rf_model.score(X_train_piecewise, y_train_piecewise))
print('The R square of the Random Forests on test set is', rf_model.score(X_test_piecewise, y_test_piecewise))
MSE = mse(y_test_piecewise, prediction)
RMSE = MSE**0.5
print("MSE:", MSE)
print("RMSE:", RMSE)
```

```python
rf_model.fit(X_train_linear, y_train_linear)
prediction_linear=rf_model.predict(X_test_linear)
print('The R square of the Random Forests on train set is', rf_model.score(X_train_linear, y_train_linear))
print('The R square of the Random Forests on test set is', rf_model.score(X_test_linear, y_test_linear))
MSE = mse(y_test_linear, prediction_linear)
RMSE = MSE**0.5
print("MSE:", MSE)
print("RMSE:", RMSE)
```
